# train_cifar10.py

**some arguments**
- --linfty: for l_infty training.
- --opt_iter 1: iterations to solve the optimization (11) in the paper approximately during training.
- --sniter 1: iterations to obtain the layerwise Lipschitz outer bounds approximately during training (In practice, 1 is enough since it is iterated for each training step).
- --test_opt_iter 10000: iterations to solve the optimization exactly during testing (it usually gets the solution in 3-4 iterations).
- --test_sniter 1000000: iterations to obtain the layerwise Lipschitz outer bounds exactly during testing. Even with a random initialization, we can computes the Lisphictz constants using the power iterations<60000. Once the Lipschitz constants are obtained, we don't have to run power iteration again.
- --prefix 'file_name_starts_with': to make the file name starts with the given value to avoid overwrting.
- --model 'c6f2': to specify a model structure.
